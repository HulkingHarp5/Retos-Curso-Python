```python
# Reto día 2

# Configurar entorno virtual (importar)
import pandas as pd
import numpy as np

# 2) Crear una Serie
calificaciones = {'Ana': 10, 'Luis': 7.5, 'Marta': 9.0, 'Juan': 5.5}

s_calificaciones = pd.Series(calificaciones)
print("\n--- Serie Original ---")
print(s_calificaciones)

# Crear nueva Serie sumando 0.5
s_calificaciones_final = s_calificaciones + 0.5
print("\n--- Serie Final (+0.5) ---")
print(s_calificaciones_final)

# Crear un DataFrame de inventario
data = {
    'producto': ['Laptop', 'Mouse', 'Teclado', 'Monitor'],
    'stock': [20, 150, 80, 35],
    'precio_usd': [1200, 25, 60, 300]
}

df_inventario = pd.DataFrame(data)
print("\n--- DataFrame de Inventario (Original) ---")
print(df_inventario)

# Hacer 'producto' el índice
df_inventario.set_index('producto', inplace=True)
print("\n--- DataFrame con Índice 'producto' ---")
print(df_inventario)

# Cargar dataset de Videojuegos
url_juegos = "https://gist.githubusercontent.com/designernatan/27da044c6dc823f7ac7fe3a01f4513ed/raw/d15b5c7d7a5efb38750b16ec935fc126ec9a6e79/vgsales.csv"

df_juegos = pd.read_csv(url_juegos)

# Flujo de Inspección

# Paso 1: Primeras 10 filas
print("\n--- Paso 1: Primeras 10 filas ---")
print(df_juegos.head(10))

# Paso 2: Forma del dataset
print("\n--- Paso 2: Forma del dataset (.shape) ---")
print(df_juegos.shape)
# Comentario: El dataset tiene df_juegos.shape[0] juegos.
# (Ejemplo: si imprime (16598, 11), significa 16,598 juegos.)

# Paso 3: info()
print("\n--- Paso 3: .info() ---")
df_juegos.info()
# Comentario: Las columnas con valores nulos son 'Year' y 'Publisher'.

# Paso 4: describe()
print("\n--- Paso 4: .describe() ---")
print(df_juegos.describe())
# Comentarios:
# - El año más reciente (max de 'Year') aparece en la estadística 'max'.
# - El promedio de ventas en Norteamérica (NA_Sales) está en la fila 'mean'.

# Paso 5: columnas
print("\n--- Paso 5: Nombres de columnas (.columns) ---")
print(df_juegos.columns)
```
